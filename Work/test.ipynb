```python
import matplotlib.pyplot as plt
import numpy as np

# A importação a seguir é necessária para a projeção '3d' funcionar.
from mpl_toolkits.mplot3d import Axes3D

# Passo 1: Crie uma figura
fig = plt.figure(figsize=(8, 12))  # Ajuste o tamanho da figura para acomodar os plots na vertical

# --- PLOT 3D (Subplot de Cima) ---
# Passo 2: Adicione o primeiro subplot com a projeção 3D
# A grade é 2 linhas, 1 coluna, e este é o primeiro subplot
ax1 = fig.add_subplot(2, 1, 1, projection='3d')

# Crie dados de exemplo para o gráfico 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plote uma superfície 3D
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='royalblue', cstride=8)
ax1.set_title('Gráfico 3D')

# --- PLOT 2D (Subplot de Baixo) ---
# Passo 3: Adicione o segundo subplot 2D
# Este é o segundo subplot da mesma grade (2 linhas, 1 coluna)
ax2 = fig.add_subplot(2, 1, 2)

# Crie dados de exemplo para o gráfico 2D
x_2d = np.linspace(0, 10, 100)
y_2d = np.sin(x_2d)

# Plote uma linha 2D
ax2.plot(x_2d, y_2d, color='red')
ax2.set_title('Gráfico 2D')
ax2.set_xlabel('Eixo X')
ax2.set_ylabel('Eixo Y')

# Ajuste o layout para evitar sobreposição
fig.tight_layout()

# Passo 4: Mostre o gráfico
fig.show()
```
